## Structured output


> #### Structured output allows agents to return data in a specific, predictable format. Instead of parsing natural language responses, you get structured data in the form of JSON objects, Pydantic models, or dataclasses that your application can use directly.

### `1. Pydantic` -> Pydantic models provide the richest feature set with field validation, descriptions, and nested structures.

- a. Using a chat model

```python
from langchain.chat_models import init_chat_model
model = init_chat_model(
    model="groq:qwen/qwen3-32b"
    
) 
model
```

```python
model.profile
```

```python
from pydantic import BaseModel,Field
class Anime(BaseModel):
    title:str = Field(description="the title of the anime")
    country:str = Field(description="Country name of origin of the anime")
    year:int = Field(description="the year in which first episode was released")
    writer:str = Field(description="the writer or author who wrote the storyline")
    theme:str = Field(description="the theme of the anime in 50-75 words,telling what the anime is about")
    rating:float = Field(description="The rating of the show/anime")
    total_episode:int = Field(description="total episode released so far")
    
```

```python
structured_model = model.with_structured_output(
    Anime,
    # include_raw=True # if you want Output message along with Structred parsing
    )
```

```python
from langchain.messages import HumanMessage,SystemMessage

messages = [
    SystemMessage("You are a anime expert and addict.you Know everything about the anime , every tiny detail of all the episodes and when some ask about anime or ask for recommandation you tell them with a crazy story-telling style "),
    HumanMessage("what is Naruto shipudin")
]
```

```python
r = model.invoke(messages)
```

```python
render_md(r.content)
```

```python
r_ = structured_model.invoke(messages)
```

```python
from rich import print as pp
pp(r_)
```

- b. Now using agent 

```python
from pydantic import Field,BaseModel
class Season(BaseModel):
    name: str = Field(description="The name of the season or movie")
    episodes: int = Field(description="Total episodes in this specific season")
```

```python
class Character(BaseModel):
    name: str = Field(description="Name of the character")
    info: str = Field(description="Short background")
    gender: str = Field(description="Gender: male, female, or other")
```

```python
class AnimeDetails(BaseModel):
    title: str = Field(description="The title of the anime")
    country: str = Field(description="Country of origin")
    year: int = Field(description="Release year")
    writer: str = Field(description="Writer/Author")
    theme: str = Field(description="Theme in 50-75 words")
    
   
    seasons: list[Season] = Field(description="List of sequel seasons/movies")
    characters: list[Character] = Field(description="List of main characters")
    
    rating: float = Field(description="Rating out of 10")
    total_episode: int = Field(description="Total episodes released so far")
```

```python
from langchain.agents import create_agent
agent = create_agent(
    model="google_genai:gemini-3-flash-preview",
    system_prompt="You are an anime expert and hardcore anime addict with total knowledge of all anime. You answer with dramatic, hype-filled storytelling, deep lore references, and emotional intensity, turning every recommendation into an unforgettable anime-style narration.",
    response_format=AnimeDetails
)
```

```python
from langchain.messages import HumanMessage
prompt = {
    "messages":[
        HumanMessage("Tell me about the classroom of elite")
    ]
}
```

```python
r1 = agent.invoke(prompt)
```

```python
res_ = r1['messages'][-1].content[0]['text']
```
