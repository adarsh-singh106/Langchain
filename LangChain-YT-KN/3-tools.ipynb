## Tools 
- Tools are simply programming functions (like Python's def function():) with defined inputs and outputs.

```python
from langchain.tools import tool

@tool
def get_weather(location: str) -> str:
    """Get the weather at a location."""
    return f"It's Raining in {location}, go kiss sun ."

tools_list = [get_weather]
model_with_tools = model.bind_tools(tools_list)  
```

### Tool Loop Execution

```python
messages = [{
    "role":"user",
    "content":"what is current weather in delhi and what is famous street food there?"
}]
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)
```

```python
argument = ai_msg.tool_calls[0]['args']["location"]
```

```python
tool_res = get_weather.invoke(argument)
tool_res
```

```python

for tool_call in ai_msg.tool_calls:
    print(tool_call)
    tool_result = get_weather.invoke(tool_call)
    print(tool_result)
    print(tool_result.content)
    messages.append(tool_result)
```

```python
final_response = model_with_tools.invoke(messages)
```
