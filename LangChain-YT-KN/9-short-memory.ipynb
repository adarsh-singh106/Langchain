```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
from IPython.display import display,Markdown
def render_md(text:str):
    return display(Markdown(text))
```

```python
short_term_memory = InMemorySaver()
```

```python
agent = create_agent(
    model="groq:qwen/qwen3-32b",
    checkpointer=short_term_memory
)
```

```python
config1={"configurable":{"thread_id":"test-1"}}
config2={"configurable":{"thread_id":"test-2"}}
```

```python
from langchain.messages import HumanMessage
checkpointer_h = {
    "messages":[
        HumanMessage("Hi My name is ash")
    ]
}
```

```python
prompt = {
    "messages":[
        HumanMessage("what is my name")
    ]
}
```

```python
res1 = agent.invoke(checkpointer_h,config=config1)
```

```python
render_md(res1['messages'][-1].content)
```

```python
res2 = agent.invoke(prompt,config1)
```

```python
render_md(res2['messages'][-1].content)
```
