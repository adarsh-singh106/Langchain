```python
from dotenv import load_dotenv
load_dotenv()
from IPython.display import display , Markdown
def md(text:str):
    return display(Markdown(text))
```

# Implementing chain using Class Modal

#### # Step-1 : Model Setup

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="openai/gpt-oss-120b")
```

#### # Step-2 : Create a prompt that expects history

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder, HumanMessagePromptTemplate
from langchain_core.messages import SystemMessage

prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="You are helpful assistant"), # Note: It's safer to specify 'content=' or just use the tuple below
        MessagesPlaceholder(variable_name='history'),
        HumanMessagePromptTemplate.from_template("{query}") # âœ… Fixed here
    ]
)
```

#### # Step-3 : Create Chain

```python
chain = prompt | model
```

#### # Step-4 Setup Memory Management

```python
from langchain_community.chat_message_histories import ChatMessageHistory
# âœ… Use the Core library (lighter and faster)
from langchain_core.chat_history import InMemoryChatMessageHistory

store = {}

def get_session_history(session_id: str):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]
```

#### # Step-5 Wrap the chain with Message History

```python
from langchain_core.runnables.history import RunnableWithMessageHistory 
model_with_memory = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="query",
    history_messages_key="history"

)
```

#### # Step-6 Making Fn to use it with ease

```python
def chat(message:str,session_id:str):
    res = model_with_memory.invoke(
        {"query":message},
        config={"configurable":{"session_id":session_id}
        }
    )
    return res.content
```
