### ðŸ§  The Concept
**Agents** combine **Language Models** with **Tools** to create intelligent systems that can:
* ðŸ¤” **Reason** about tasks.
* ðŸ› ï¸ **Decide** which tools to use.
* ðŸ”„ **Iteratively work** towards solutions.

---

### âš™ï¸ Implementation
`create_agent` provides a **production-ready** agent implementation.

> **How it works:**
> An LLM Agent runs tools in a **loop** to achieve a goal. The agent runs until a **stop condition** is met (e.g., the model emits a final output or an iteration limit is reached).

---

### ðŸ”„ The Agent Loop (Workflow)



The agent operates in a continuous cycle of **Action** and **Observation**:

1.  **ðŸ“¥ Input** â†’ The user provides a goal.
2.  **ðŸ¤– Model** â†’ The brain decides what to do next.
3.  **âš¡ Action** â†’ The agent calls a specific tool.
4.  **ðŸ› ï¸ Tools** â†’ The tool executes and returns data.
5.  **ðŸ‘€ Observation** â†’ The agent sees the result of the tool.
6.  **ðŸ Finish** â†’ The model produces the **Output** when done.

```python
from langchain.tools import tool
@tool
def get_weather(location:str)->str:
    """it gives weather for given location"""
    return f"its raining in {location}."
```

```python
from langchain.agents import create_agent
agent = create_agent(
    model="groq:qwen/qwen3-32b", # In Agent there is not model_provider params so write `groq:` before model name
    tools=[get_weather],
    # system_prompt="You are very helpful assisant and talks in poetic manner!"
)
```

```python
prompt = {
    "messages":[{
        "role":"user",
        "content":"what is weather in delhi?"
    }]
}
```

```python
res = agent.invoke(prompt)
```

```python
render_md(res['messages'][-1].content)
```
