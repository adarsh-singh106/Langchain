# 1. Google Gemini

### - Calling Gemini using `init_chat_model`

```python
from langchain.chat_models import init_chat_model;
gemini_model = init_chat_model("google_genai:gemini-2.5-flash")
```

```python
prompt = [
    ("user","what is 8+9 then divided by 17?")
]
```

```python
res = gemini_model.invoke('what is 8+9 then divided by 17?')
```

```python
from IPython.display import Markdown,display
def render_md(text:str):
    return display(Markdown(text))
```

```python
render_md(res.content)
```

### - Caling gemini using `langchain_google_genai (Modal Class)`

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
res_ = llm_gemini.invoke(prompt)
```

```python
render_md(res_.content)
```
